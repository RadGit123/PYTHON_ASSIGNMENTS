**1.DISTRIBUTION  OF EMPLOYEES ACROSS THE TEAMS.**

```python

NumberofEmployees = df["Team"].value_counts()

EmployeePercentage = (NumberofEmployees/ df.shape[0]) * 100

EmployeeData = pd.DataFrame({"Team": NumberofEmployees .index, "Count": NumberofEmployees .values, "Percentage": EmployeePercentage.values})
print(EmployeeData)


plt.figure(figsize=(6, 4))
sns.barplot(x=NumberofEmployees.index, y=NumberofEmployees.values, palette="gray")
plt.xlabel("Team")
plt.ylabel("Number of Employees")
plt.title("Employee Distribution Across Teams")
plt.xticks(rotation=90,fontsize=6)
plt.yticks(rotation=90,fontsize=6)
plt.show()

plt.figure(figsize=(8, 8))
plt.pie( EmployeePercentage.values,labels=NumberofEmployees.index,autopct='%1.1f%%',startangle=120,textprops={'fontsize': 6})
plt.title("Employee Percentage distribuition")
plt.axis("equal",)
plt.show()
```

**Team wth highest number of employee:New Orleans Pelicans**

**2.SEGREGATION BASED ON POSITION.**

3. **PREDOMINENT AGE GROUP.**

```python

Bounds = [18, 25, 35, 45, 55, 65]
Agelabels = ["18-24", "25-34", "35-44", "45-54", "55-64"]


df["Age Group"] = pd.cut(df["Age"], bins=Bounds, labels=Agelabels, right=False)

AgeWiseEmployees = df["Age Group"].value_counts()
print(AgeWiseEmployees)

EmployeeAgeData = pd.DataFrame({"Age Group":AgeWiseEmployees .index, "Count": AgeWiseEmployees .values,})
print(EmployeeAgeData )

plt.figure(figsize=(6, 4))
sns.barplot(x=AgeWiseEmployees.index, y=AgeWiseEmployees .values, palette="pastel")
plt.xlabel("Age Group")
plt.ylabel("Number of Employees")
plt.title("Predominant Age Group")
plt.show()
```

**Most predominent age group:25-34**

 **4.TEAM AND POSITION WITH THE HIGHEST SALARY EXPENDITURE.-SEPEARATELY**.

```python

Teamsalary = df.groupby("Team")["Salary"].sum().sort_values(ascending=False)
print(Teamsalary)


plt.figure(figsize=(6, 4))
sns.barplot(x=Teamsalary.index, y=Teamsalary.values, palette="cividis")
plt.xlabel("Team")
plt.ylabel("Total Salary Expenditure")
plt.title("Total Salary Expenditure by Team")
plt.xticks(rotation=90,fontsize=6)
plt.yticks(rotation=90,fontsize=6)
plt.show()


Positionsalary = df.groupby("Position")["Salary"].sum().sort_values(ascending=False)
print(Positionsalary)


plt.figure(figsize=(6, 4))
sns.barplot(x=Positionsalary.index, y=Positionsalary.values, palette="Blues_r")
plt.xlabel("Position")
plt.ylabel("Total Salary Expenditure")
plt.title("Total Salary Expenditure by Position")
plt.xticks(rotation=90)
plt.show()
```

**Team with highest salary expenditure:Cleveland Cavaliers.**
**Position with highest salary expenditure:C.**

**METHOD 2-TEAM & POSITION WITH HIGHEST SALARY.**

```python

TeamPositionsalary = df.groupby(['Team', 'Position'])['Salary'].sum().sort_values(ascending=False)

print(TeamPositionsalary.head(30))

highest_salary_expenditure_pair = TeamPositionsalary.idxmax() 
highest_expenditure_value = TeamPositionsalary.max()  


print(f"The team and position with the highest salary expenditure is: {highest_salary_expenditure_pair}")
print(f"The total salary expenditure for this team and position is: {highest_expenditure_value}")

plt.figure(figsize=(20,10))
sns.barplot(x=team_position_salary.index.map(lambda x: f"{x[0]} - {x[1]}"), y=team_position_salary.values, palette="magma")
plt.xlabel("Team - Position")
plt.ylabel("Total Salary Expenditure")
plt.title("Total Salary Expenditure by Team and Position")
plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

**Among all the teams and positions the "SF" position from " Los Angeles Lakers"  team has highest sum of salary.**

**5.CORRELATION BETWEEN AGE AND SALARY**

```python

correlation = df["Age"].corr(df["Salary"])
print(f"Correlation between Age and Salary: {correlation}")

plt.figure(figsize=(6, 4))
sns.scatterplot(x=df["Age"], y=df["Salary"], alpha=0.6, color="red")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age & Salary Correlation")
plt.show()
```

**There is a postive correlation of age and salary till the age group 30,but after that there is a fluctuation.**
